# Fixing some details

- As we can see, we have an extra column named 'Unnamed'. So we need to drop it

```python
movies = movies.drop(columns=['Unnamed: 0'])
```

```python
movies.head(1)
```

# Creating a new fan rating
To a better metric to evaluate the quality of the movies, we create a new column with the average of the persons grade

```python
movies.iloc[:,11:]
```

```python
movies.iloc[:,11:].mean()
```

```python
# Creating the new column Fans which will have the average of each column
movies['Fans'] = movies.iloc[:,11:].mean(axis=1)

movies.head()
```

# Dropping columns with Null values

- As we've seen before, there is presence of missing values in three columns of the dataset. So we need to drop it in order to not affect the machine learning model

```python
movies.isnull().sum()
```

```python
# We eliminate the columns with missing values
movies.dropna(inplace = True)
```

Detecting outliers

- We need to check out if there are outliers in our dataset, so we proceed to create a boxplot to detect this kind of values

```python
fig, axes = plt.subplots(3, 3, figsize = (10, 10))
fig.tight_layout(pad=3)
sns.boxplot(x=movies['Personalrating'], ax=axes[0,0])
sns.boxplot(x=movies['Fans'], ax=axes[0,1])
sns.boxplot(x=movies['Duration'], ax=axes[0,2])
sns.boxplot(x=movies['Budget'], ax=axes[1,0])
sns.boxplot(x=movies['Revenue'], ax=axes[1,1])
sns.boxplot(x=movies['Imdb'], ax=axes[1,2])
sns.boxplot(x=movies['Critic_1'], ax=axes[2,0])
sns.boxplot(x=movies['Critic_2'], ax=axes[2,1])
sns.boxplot(x=movies['Critic_3'], ax=axes[2,2])
```

```python
movies_numeric = movies.select_dtypes(include='number')
```

```python
Q1 = movies_numeric.quantile(0.25)
Q3 = movies_numeric.quantile(0.75)
IQR = Q3 - Q1
Q1
```

```python
movies[((movies_numeric < (Q1 - 1.5 * IQR)) |(movies_numeric > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
movies_duration = movies['Duration']
movies_duration[((movies_duration < (Q1['Duration'] - 1.5 * IQR['Duration'])) |(movies_duration > (Q3['Duration'] + 1.5 * IQR['Duration'])))]
```

```python
movies_budget = movies['Budget']
budget_outliers = movies_budget[((movies_budget < (Q1['Budget'] - 1.5 * IQR['Budget'])) |(movies_budget > (Q3['Budget'] + 1.5 * IQR['Budget'])))]
budget_outliers
```

```python
movies.drop(budget_outliers.index,inplace=True)
movies.shape
```

```python
fig, axes = plt.subplots(3, 3,figsize=(10,10))
fig.tight_layout(pad=3)
sns.boxplot(x=movies['Personalrating'], ax=axes[0,0])
sns.boxplot(x=movies['Fans'], ax=axes[0,1])
sns.boxplot(x=movies['Duration'], ax=axes[0,2])
sns.boxplot(x=movies['Budget'], ax=axes[1,0])
sns.boxplot(x=movies['Revenue'], ax=axes[1,1])
sns.boxplot(x=movies['Imdb'], ax=axes[1,2])
sns.boxplot(x=movies['Critic_1'], ax=axes[2,0])
sns.boxplot(x=movies['Critic_2'], ax=axes[2,1])
sns.boxplot(x=movies['Critic_3'], ax=axes[2,2])
```

```python
# Eliminate the revenue outliders
movies_revenue = movies['Revenue']
revenue_outliers = movies_revenue[((movies_revenue < (Q1['Revenue'] - 1.5 * IQR['Revenue'])) |(movies_revenue > (Q3['Revenue'] + 1.5 * IQR['Revenue'])))]
revenue_outliers
```

```python
movies.drop(revenue_outliers.index,inplace=True)
movies.shape
```

```python
movies_imdb = movies['Imdb']
imdb_outliers=movies_imdb[((movies_imdb < (Q1['Imdb'] - 1.5 * IQR['Imdb'])) |(movies_imdb > (Q3['Imdb'] + 1.5 * IQR['Imdb'])))]
imdb_outliers
```

```python
movies.drop(imdb_outliers.index,inplace=True)
movies.shape
```

```python
movies_critic1 = movies['Critic_1']
critic1_outliers = movies_critic1[((movies_critic1 < (Q1['Critic_1'] - 1.5 * IQR['Critic_1'])) |(movies_critic1 > (Q3['Critic_1'] + 1.5 * IQR['Critic_1'])))]
critic1_outliers
```

```python
movies.drop(critic1_outliers.index,inplace=True)
movies.shape
```

```python
movies_fans = movies['Fans']
fans_outliers = movies_fans[((movies_fans < (Q1['Fans'] - 1.5 * IQR['Fans'])) |(movies_fans > (Q3['Fans'] + 1.5 * IQR['Fans'])))]
fans_outliers
```

```python
movies.drop(fans_outliers.index,inplace=True)
movies.shape
```

```python
movies_critic3 = movies['Critic_3']
critic3_outliers = movies_critic3[((movies_critic3 < (Q1['Critic_3'] - 1.5 * IQR['Critic_3'])) |(movies_critic3 > (Q3['Critic_3'] + 1.5 * IQR['Critic_3'])))]
critic3_outliers
```

```python
movies.drop(critic3_outliers.index,inplace=True)
movies.shape
```

```python
fig, axes = plt.subplots(3, 3,figsize=(10,10))
fig.tight_layout(pad=3)
sns.boxplot(x=movies['Personalrating'], ax=axes[0,0])
sns.boxplot(x=movies['Fans'], ax=axes[0,1])
sns.boxplot(x=movies['Duration'], ax=axes[0,2])
sns.boxplot(x=movies['Budget'], ax=axes[1,0])
sns.boxplot(x=movies['Revenue'], ax=axes[1,1])
sns.boxplot(x=movies['Imdb'], ax=axes[1,2])
sns.boxplot(x=movies['Critic_1'], ax=axes[2,0])
sns.boxplot(x=movies['Critic_2'], ax=axes[2,1])
sns.boxplot(x=movies['Critic_3'], ax=axes[2,2])
```
